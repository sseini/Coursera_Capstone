## Segmenting and Clustering Neighborhoods in Toronto

```python
import pandas as pd
import numpy as np
```

#### Using pandas.read_html to read the wikipedia webpage https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M

```python
df_pc = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
```

#### Deleting the rows witch contains NaN value

```python
df_pc=df_pc[df_pc["Borough"] != 'Not assigned'].dropna()
```

#### Replace the "Code postal" label by "PostalCode" 

```python
df_pc=df_pc.rename(columns={"Postal Code": "PostalCode"})
df_pc=df_pc.replace(to_replace="\/", value=",", regex=True)
if "PostalCode" in df_pc.columns: 
    df_pc.set_index(["PostalCode"], drop=True, inplace=True)
```
